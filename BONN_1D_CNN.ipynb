```python
import numpy as np
import h5py as h5
```

```python
dataset_relative_path='dataset/random-iter-1b/'
```

```python
datafile=dataset_relative_path+'datafile.h5'
```

```python
import h5py as h5
with h5.File(datafile, 'r') as datafile:
    X_train = np.array(datafile['X_train'])
    Y_train = np.array(datafile['Y_train'])
    
    X_test = np.array(datafile['X_test'])
    Y_test = np.array(datafile['Y_test'])
```

```python
def make_dimensions_compatible(arr):
    return arr.reshape(arr.shape[0],-1,1)
```

```python
X_train = make_dimensions_compatible(X_train)
X_test = make_dimensions_compatible(X_test)
```

```python
X_train.shape,X_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv1D, Dropout, MaxPooling1D
```

```python
model = Sequential()
model.add(Conv1D(filters=4, kernel_size=6, activation='relu', input_shape=(4097,1)))#1
model.add(MaxPooling1D(pool_size=2))#2
model.add(Conv1D(filters=4, kernel_size=5, activation='relu'))#3
model.add(MaxPooling1D(pool_size=2))#4
model.add(Conv1D(filters=10, kernel_size=4, activation='relu'))#5
model.add(MaxPooling1D(pool_size=2))#6
model.add(Conv1D(filters=10, kernel_size=4, activation='relu'))#7
model.add(MaxPooling1D(pool_size=2))#8
model.add(Conv1D(filters=15, kernel_size=4, activation='relu'))#9
model.add(MaxPooling1D(pool_size=2))#10
model.add(Flatten())#11
#model.add(Dropout(0.5))
#model.add(MaxPooling1D(pool_size=2))
#model.add(Flatten()) model.add(LeakyReLU(alpha=0.05))
model.add(Dense(50, activation='relu'))#12
model.add(Dense(20, activation='relu'))#13
model.add(Dense(3, activation='softmax'))#14
print(model.summary())
```

```python
epochs=150
batch_size=3
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history=model.fit(X_train, Y_train, epochs=epochs, batch_size=batch_size)
	# evaluate model
_, accuracy = model.evaluate(X_test, Y_test, batch_size=batch_size)
```
